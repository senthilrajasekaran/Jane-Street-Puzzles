```python
def multinomial(race):
    tot =  40320
    for i in range(24):
        tot = tot // factorial(race.count(i))
    return tot
    

def factorial(n):
    if n==0:
        return 1
    return n* factorial(n-1)

def choose(n,k):
    return factorial(n)//(factorial(k)*factorial(n-k))


def get_prob(race,p):
    
    nr = sum(race)
    prob = ((1-p)**(23-nr)) * (p**nr) * choose(23,nr)
    
    return prob

def binom(trials, success, prob):
    return choose(trials,success) * (prob**(success)) * ( (1-prob)**(trials-success))


def prob_race(race):
    
    first = race[0:4]
    second = race[4:8]
    
    trials = sum(race)
    success = sum(first)
    
    prob = binom(trials,success,.5)
    
    
    a = race[0:2]
    b = race[2:4]
    c = race[4:6]
    d = race[6:8]
    
    trials = sum(a)+sum(b)
    success = sum(a)
    
    prob = prob * binom(trials,success,.5)
    
    trials = sum(c)+sum(d)
    success = sum(c)
    
    prob = prob * binom(trials,success,.5)
    
    
    
    i=0
    while i<=7:
        pair = race[i:i+2]
        trials = sum(pair)
        success = pair[0]
        
        prob = prob * binom(trials,success,.5)
        i = i+2
        
    return prob

    
    
    
    
```

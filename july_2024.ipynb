```python
def word_reverses(word, length):
    reverse_word = word[::-1]
    index = 0
    while index + length < len(word) +1:
        candidate = reverse_word[index:index+length]
        if candidate in word_list:
            
            print (word + " contains " + candidate)
        index += 1 
        
    
```

```python
def find_reverse(s,length):
    words = [w for w in word_list if len(w) == length]
    
    for word in words:
        reverse_word = word[::-1]
        index = 0
        while index + len(s) < len(word) +1:
            candidate = reverse_word[index:index+len(s)]
            if candidate == s:
            
                print (word + " contains " + candidate)
            index += 1 
        
        
```

```python
def step(l):
    # l = [a,b,c,d]
    a = abs(l[0] - l[1])
    b = abs(l[1] - l[2])
    c = abs(l[2] - l[3])
    d = abs(l[3] - l[0])
    
    return [a,b,c,d]

def check(l):
    return all(item == 0 for item in l)

def f(l,verbose = False):
    i = 1
    while not check(l):
        if verbose and almost(l):
            print(l)
        l = step(l)
        i += 1
    return i
    
def almost(l):
    return l[0] == l[1] and l[2] == l[3] and l[0] == l[2] and l[0] > 0
    
```

```python
def looper(bound):
    m = 0
    best = []
    for a in range(bound):
        for b in range(0,a+1):
            for c in range(0, b+1):
                for d in range(0, c+1):
                    if f([a,b,c,d]) == m:
                        best.append([a,b,c,d])
                    if f([a,b,c,d]) > m:
                        best = [[a,b,c,d]]
                        m = f([a,b,c,d])

    return best
            
```

```python
def choose(i,k):
    
    if k == 0:
        return [[]]
    else:
        prev = choose(i, k-1)
        new = []
        for x in prev:
            m = 0
            if x != []:
                m = max(x)
            if m == i:
                break
            
            for j in range(m,i):
                if j not in x:
                    new.append( x + [j])
        return new

    
def make_index(i):
    return choose(i+3,3)

def make_partitions(i):
    partitions = []
    for j in make_index(i):
        a = j[0]
        b = (j[1]-j[0])-1
        c = (j[2]-j[1])-1
        d = ( (i+3) - j[2]) - 1
        #print( str(j) + " : "  + str([a,b,c,d]) )
        partitions.append([a,b,c,d])
    return partitions
        
    
def maximumtwo(bound):
    m = -1
    best = []
    for a in make_partitions(bound):
        if f(a) == m:
            best.append(a)
        if f(a) > m:
            best = [a]
            m = f(a)

    return [m,best]

def maximumthree(bound):
    m = -1
    for a in make_partitions(bound):
        if f(a) > m:
            m = f(a)

    return m
    
```

```python
def trib(n):
    if n == 0:
        return 0
    if n == 1:
        return 0
    if n == 2:
        return 1
    else:
        return trib(n-1) + trib(n-2) + trib(n-3)
```

```python
def biggest(b):
    i=0
    while trib(i) <= b:
        i+= 1
        
    return [i,trib(i-1),trib(i-2),trib(i-3),trib(i-4)]
```
